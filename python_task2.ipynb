1.Given a sentence, identify all unique words and display:
The count of unique words
The unique words themselves

Input
Enter sentence: python is easy and python is powerful

Expected Output
Unique words count: 5  
Unique words: {'python', 'is', 'easy', 'and', 'powerful'}


```python
Sent = input("Enter Sentence")
word = set()
l = Sent.split()

for i in l:
    word.add(i)
print(f"Unique words count: {len(word)}")
print(f'Unique words {word}')
```

2.Store employee names and their salaries, and determine which employee has the highest salary.

Input Format
A dictionary where:
Keys represent employee names
Values represent employee salaries
Example Input
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

Expected Output
Highest Salary: Ravi - 75000


```python
dist = { "Ravi": 75000, "Anita": 68000, "Kiran": 72000 }
high = 0
name = ""

for i, j in dist.items():
    if j > high:
        high = j
        name = i

print(f"Highest Salary {name} - {high}")

```

```python
d = sorted(dist.items(),key=lambda x:x[1], reverse=True)
print(f"Highest Salary {d[0][0]} - {d[0][1]}")
```

3.Find the maximum and minimum values from a list of numbers.

Input Format
A list of integers where each element represents a numeric value.
Example Input
numbers = [45, 22, 89, 10, 66]

Expected Output
List: [45, 22, 89, 10, 66]  
Max: 89  
Min: 10


```python
l = [45, 22, 89, 10, 66]
h = l[0]
s = l[0]
for i in l:
    if i > h:
        h = i
    if i < s:
        s = i
print(f'max {h}')
print(f'min {s}')
```

Count how many products have a price greater than 1000.

Input Format
A list of integers where each element represents the price of a product.
Example Input
prices = [450, 1200, 899, 1500, 300]

Expected Output
Products above 1000: 2



```python
prices = [450, 1200, 899, 1500, 300]
count = 0
for i in prices:
    if i > 1000:
        count +=1
print(f'Products above 1000: {count}')
```

5.Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).

Example Input
attendance = ["P", "P", "A", "P", "P"]

Expected Output
Attendance Percentage: 80.0


```python
attend = ["P", "P", "A", "P", "P"]
c = attend.count('P')
total = len(attend)
perc = (c/total)*100
print(f'Attendance Percentage:{perc}')
```
